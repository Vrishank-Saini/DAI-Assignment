# Bivariate Analysis


```python
df = df.loc[:, ['Population (in Millions)', 'GDP (in Trillions USD)']]
```

```python
# Checking for duplicates and missing values
print(df.duplicated())
print(df.isna().sum())

df.describe()
```

```python
df.corr()
```

```python
#Anova Testing
f_statistic, p_value = stats.f_oneway(df['Population (in Millions)'], df['GDP (in Trillions USD)'])
print('F statistic:', f_statistic)
print('P value:', p_value)
# p value is less than 0.05, we reject the null hypothesis that the two columns are independent
# The two columns depend upon each other.
```

```python
sns.scatterplot(x="Population (in Millions)",y="GDP (in Trillions USD)",data=df)
```

##### The chart illustrates that population size is not directly proportional to GDP. While countries with larger populations like China have substantial GDPs, others, like India, have lower GDP despite similar population sizes. Conversely, the USA has a much higher GDP relative to its population size.

```python
plt.figure(figsize=(10,12))
sns.barplot(x="Country",y="Military Expenditure (in Billion USD)",data=df)
plt.title("Military Expenditure by each country")
plt.show()



plt.figure(figsize=(10,12))
sns.barplot(x="Country",y="Smartphone Adoption (%)",data=df,palette="Set2")
plt.title("Smartphone Adoption in each country")
plt.show()
```

####  Conclusion: 1)USA spends most on Military expenditure followed by China, India and Russia.

```python
sns.scatterplot(x="Life Expectancy (Years)",y="Healthcare Expenditure per Capita (USD)",data=df)
plt.title("Life Expectancy vs Healthcare Expenditure per Capita (USD)")
plt.show()


sns.scatterplot(x="Urban Population (%)",y="Population Growth Rate (%)",data=df)
plt.title("Urban Population vs Population Growth Rate")
plt.show()
```
